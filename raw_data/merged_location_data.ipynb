```python
locations =  pd.read_csv("locations.csv")
locations.drop(['Alpha-2 code', 'Alpha-3 code', 'Numeric code', 'Icon'], axis =1, inplace = True)
locations.rename(columns={'Country': 'Country Name'}, inplace=True)
locations.head()
```

```python
battle_deaths_by_country = pd.read_csv("battle_deaths_by_country.csv")
battle_deaths_by_country
```

```python
battle_deaths_by_country = pd.read_csv("battle_deaths_by_country.csv")
battle_deaths_by_country_merged = pd.merge(locations, battle_deaths_by_country, on='Country Name')
battle_deaths_by_country_merged
```

```python
battle_deaths_by_country_merged.to_csv("battle_deaths_by_country_merged.csv", encoding='utf-8', index=False)
```
